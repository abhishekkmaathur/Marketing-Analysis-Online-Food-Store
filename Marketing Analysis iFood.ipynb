### Clustering - K-Means

```python
#Checking the mean and std. We can notice variations in the dataset
df.describe().T
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale_df = df.drop(['age_group','age_band'], axis =1)
```

```python
scale_df.head()
```

```python
X = StandardScaler()
```

```python
scaled_df = X.fit_transform(scale_df)
```

```python
scaled_df
```

```python
#We will identify the nmbers of clusters we need to run in K means. We will find via WSS (Within Sum of Squares) plot
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
```

```python
wss =[]
```

```python
for i in range(1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
```

```python
# We can see that the drop in wss is not significant after 2. We can say 3 will be the optimal numbers of clusters.
```

```python
k_means = KMeans(n_clusters = 2)
k_means.fit(scaled_df)
labels = k_means.labels_
```

```python
labels
```

```python
df['clus_kmeans']= labels
df.head()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(scaled_df, labels)
```

### Prediction with Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import roc_auc_score,roc_curve,classification_report,confusion_matrix,plot_confusion_matrix
```

```python
df_model = pd.read_csv('ifood_df.csv')
```

```python
df_model.head()
```

```python
X= df_model.drop('Response', axis =1)
y =  df_model.Response
```

```python
X_train_RF,X_test_RF,y_train_RF,y_test_RF= train_test_split(X,y,test_size=.30,random_state=0)
```

```python
RF_model=RandomForestClassifier(n_estimators=100,random_state=1)
RF_model.fit(X_train_RF,y_train_RF)
```

```python
#Training Performance
y_train_predict_RF= RF_model.predict(X_train_RF)
RF_model_score= RF_model.score(X_train_RF,y_train_RF)
print(RF_model_score)
```

```python
#Classification Report and Confusion Matrix on Train

print(classification_report(y_train_RF, y_train_predict_RF))
fig,ax = plt.subplots(figsize = (8,6))
plot_confusion_matrix(RF_model, X_train_RF,y_train_RF, ax = ax);
plt.title('RF-Train Confusion Matrix')
```

```python
#Test Performance
y_test_predict_RF= RF_model.predict(X_test_RF)
RF_model_score= RF_model.score(X_test_RF,y_test_RF)
print(RF_model_score)
```

```python
#Classification Report and Confusion Matrix on Test Data

print(classification_report(y_test_RF, y_test_predict_RF))
fig,ax = plt.subplots(figsize = (8,6))
plot_confusion_matrix(RF_model, X_test_RF,y_test_RF, ax = ax);
plt.title('RF-Train Confusion Matrix')
```

```python
#Random Forest with Hyperparameters Tunning
param_grid = {'max_depth': [5,6,7],
    'max_features': ['auto'],
    'min_samples_leaf': [20,25,30],
    'min_samples_split': [50,60,70],
    'n_estimators': [50,100,150]
}

rfcl = RandomForestClassifier(random_state=1)
```

```python
grid_search_RF=GridSearchCV(estimator = rfcl, param_grid = param_grid,cv = 5)
```

```python
grid_search_RF.fit(X_train_RF,y_train_RF)
```

```python
grid_search_RF.best_params_
```

```python
best_grid_RF = grid_search_RF.best_estimator_
```

```python
best_grid_RF
```

```python
ytrain_predict_RF = best_grid_RF.predict(X_train_RF)
ytest_predict_RF = best_grid_RF.predict(X_test_RF)
```

```python
#Confusion Matrix on Train

confusion_matrix(y_train_RF,ytrain_predict_RF)
```

```python
#Confusion Matrix on Text
confusion_matrix(y_test_RF,ytest_predict_RF)
```
